# Hidden rewards

```python
# Example usage
depth = 4
branching_factor = 2
discount_factor = 0.8
n = 4
tree = Tree(depth, branching_factor, n, discount_factor, make_viz=True, sampling="uniform", randomize=False, reward_bound=0.8)
```

```python
tree.root.children[0].r = 0
tree.root.children[0].children[0].r = 0.64
tree.root.children[0].children[1].r = 0.64
tree.root.children[0].children[0].children[0].r = 0.51
tree.root.children[0].children[0].children[1].r = 0.51
tree.root.children[0].children[1].children[0].r = 0.51
tree.root.children[0].children[1].children[1].r = 0.51

tree.root.children[1].r = 0.3
tree.root.children[1].children[0].children[0].r = 0
tree.root.children[1].children[0].children[1].r = 0
tree.root.children[1].children[1].children[0].r = 0
tree.root.children[1].children[1].children[1].r = 0
```

```python
tree_copy = deepcopy(tree)
```

```python
tree_copy.uniform_search()
```

```python
graph = tree_copy.to_graphviz()
```
